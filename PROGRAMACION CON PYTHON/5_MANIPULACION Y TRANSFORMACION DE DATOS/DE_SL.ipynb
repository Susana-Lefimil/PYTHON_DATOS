```python
#Se genera una tabla con la unión de las 3 tablas. hint: utiliza sufijos para para las
#columnas que se llaman igual usando el parámetro suffixes de pd.merge()

#primero se cruza df_incidents y df_officers, se utiliza outer para no dejar ningun resgistro fuera y case_number será el id

df_merge1 = df_incidents.merge(df_officers, on='case_number', how='outer', suffixes=('_i','_o'))
df_merge1.head()
```

```python
df_merge1.shape
```

```python
#ahora con df_subjects / se observa que solo existen columnas con el mismo nnombre en df_officers y df_subjects
# por ello se utiliza sufijos o y s
df_merge2 = df_merge1.merge(df_subjects, on='case_number', how='outer', suffixes=('_o','_s'))
df_merge2.head()
```

```python
df_merge2.shape
```

```python
#Se verifica si hay filas duplicadas
df_merge2[df_merge2.duplicated(keep=False)]
```

```python
#no hay duplicados considerando todas las columnas. Ahora se considera solo la variable id "case_number"
df_merge2[df_merge2.duplicated(subset = ["case_number"])]
```

```python
#ahora basado en el id pero verificando cada data frame por separado (esto se realiza ya que no existe mucha información sobre 
# la data que se está trabajando)

df_incidents[df_incidents.duplicated(subset = ["case_number"])]
```

```python
duplicate1 = df_officers[df_officers.duplicated("case_number")]
duplicate1.sort_values(by=['case_number'])
```

```python
#tabla pivote que muestra el número de casos por mes en la fila y por año en las columnas.
#¿Se ve alguna relación entre mes del año y cantidad de casos?
# se generan columnas mes y año

df_merge2['year'] = pd.to_datetime(df_merge2['date']).dt.strftime('%Y')
df_merge2['month'] = pd.to_datetime(df_merge2['date']).dt.strftime('%m')

df_pivot_reporte1 = pd.pivot_table(
                                   data = df_merge2 # DataFrame.
                                 , index = 'month'
                                 , columns='year'
                                 , values = 'case_number'
                                 , aggfunc= "count" # función a aplicar a la columna de valores.
                                 , margins = True # Generar una fila/columna de totales.
)

df_pivot_reporte1
```

```python
#2015 es el año con más casos, mientras que el mes 3 y 6 es el que tiene mayor cantidad de casos. 
```

```python
#Tabla pivote que muestre en las filas el género del oficial y en las columnas el
#género del subject. ¿Cómo interpretas los valores que muestra esta vista?

df_pivot_reporte2 = pd.pivot_table(
                                   data = df_merge2 # DataFrame.
                                 , index = 'gender_o'
                                 , columns='gender_s'
                                 , values = 'case_number'
                                 , aggfunc= "count" # función a aplicar a la columna de valores.
                                 , margins = True # Generar una fila/columna de totales.
)

df_pivot_reporte2

```

```python
#Carga de datos de Cleaned_DS_Jobs.csv

df = pd.read_csv("Cleaned_DS_Jobs.csv")
df.head()
```

```python
df.shape
```

```python
#Eliminar nulos "na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"

df=df.replace({"na": np.nan})
df=df.replace({"NA": np.nan})
df=df.replace({-1: np.nan})
df=df.replace({"0": np.nan})
df=df.replace({"-1": np.nan})
df=df.replace({"null": np.nan})
df=df.replace({"n/a": np.nan})
df=df.replace({"N/A": np.nan})
df=df.replace({"NULL": np.nan})
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
660-80  # se borran 580 filas
```

```python
#Salario Estimado Mínimo y Máximo
df['Salary Estimate'].value_counts()
```

```python

df['Salario Estimado Mínimo']=df['Salary Estimate'].apply(lambda x: x.split('-')[0])
df['Salario Estimado Máximo']=df['Salary Estimate'].apply(lambda x: x.split('-')[1])
```
