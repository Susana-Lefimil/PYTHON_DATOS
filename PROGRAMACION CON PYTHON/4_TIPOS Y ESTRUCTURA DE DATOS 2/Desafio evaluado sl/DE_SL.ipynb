```python
import pandas as pd
import numpy as np
```

```python
#Crea un DataFrame con el archivo entregado
df = pd.read_csv('resumen_resultados_astronautas.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
#peso está en kilos y altura en cm (se pasan a metros), para ello se modifica columna altura
df['altura']=df['altura']/100
df.head()
```

```python
#Se crea columna imc

df['IMC']=df['peso']/(df['altura']*df['altura'])
```

```python
df.head()
```

```python
#Se crea columna promedio_evaluaciones


df['promedio_evaluaciones']=(df['evaluacion_fisica']+df['evaluacion_adaptativa']+df['evaluacion_psicologica']+df['evaluacion_medica'])/4
```

```python
df.head()
```

```python
#atronautas con imc entre 18.5 y 24.9
#El promedio de las 4 evaluaciones debe ser mayor o igual a 87

df=df.loc[(df['IMC']>=18.5) & (df['IMC']<=24.9) & (df['promedio_evaluaciones']>=87)]
df.describe()
```

```python
#Reinicia el índice del DataFrame resultante

df.head()
```

```python
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
#Modifica el valor de la columna "califica", para que diga "Si" en todas las filas del DataFrame resultante
df['califica']='Si'
df['califica'].value_counts()
```

```python
df.to_csv('astronautas_calificados.csv', sep=';')
```
